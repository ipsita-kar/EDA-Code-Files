```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
numeric_types=df.select_dtypes(include=[int,float]).columns
df[numeric_types].boxplot()
plt.xticks(rotation=90)
plt.title("boxplot of income")
plt.show()
```

```python
income_mean=df["Income"].mean()
df["Income"]=df["Income"].fillna(income_mean)
q1=df["Income"].quantile(0.25)
q3=df["Income"].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df["Income"]=df["Income"].apply(lambda x:income_mean if x <lower_bound or x> upper_bound else x)
```
